## Hierarchical Clustering
Hierarchical clustering is a method of grouping similar objects into clusters. It starts with each object in its
own cluster and then merges or splits clusters based on their similarity. There are two main types of
hierarchical clustering: agglomerative and divisive.It is also used to find clusters in data which is not normally distributed.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs

#Generate a random dataset with 100 samples and 3 features
X, y = make_blobs(n_samples=100, centers=3, n_features=3, random_state=0)
# Perform hierarchical clustering using the AgglomerativeClustering class
clustering = AgglomerativeClustering(n_clusters=8, metric='euclidean', linkage='ward')
clustering.fit(X)
# Create a DataFrame from the clustering results
# df = pd.DataFrame(clustering.labels_, columns=['cluster'])
# df['x'] = X[:, 0]
# df['y'] = X[:, 1]
# df['z'] = X[:, 2]
# # Create a scatter plot of the clusters
# plt.figure(figsize=(10, 6))
# sns.scatterplot(x='x', y='y', hue='cluster', data=df)
# plt.title('Agglomerative Clustering Results')
# plt.xlabel('Feature 1')
# plt.ylabel('Feature 2')
# plt.legend(title='Cluster')
# plt.show()
# Create a dendrogram to visualize the hierarchical clustering
plt.figure(figsize=(10, 6))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```
