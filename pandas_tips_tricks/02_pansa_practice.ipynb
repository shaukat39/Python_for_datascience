## Splitting a string into multiple columns

```python
import pandas as pd
```

```python
df=pd.DataFrame({'name':['Muhammad Ali','Raza Khan','Sajjid Shah','Abu Bakar'],'location':['Lahore, Pakistan','Peshawar, Pakistan','Rawalpindi, Pakistan','Multan, Pakistan']})
```

```python
df
```

```python
df.name.str.split(' ').head()
```

```python
#Splitting a column into two cloumns
df.name.str.split(' ',expand=True)
```

```python
# Adding those splits into new columns
df[['first_name','last_name']]=df.name.str.split(' ',expand=True)
```

```python
df
```

```python
df[['city','country']]=df.location.str.split(', ',expand=True)
```

```python
df
```

```python
# Refine data manipulation
df=df[['first_name','last_name','city','country']]
```

```python
df
```

## Reshaping a dataframe

```python
import numpy as np
import pandas as pd
```

```python
fasla=pd.DataFrame([['12345',100,200,300],['34567',400,500,600],['67890',700,800,900]],columns=['zip','factory','warehouse','retail'])
fasla.head()
```

```python
fasla.melt(id_vars='zip',var_name='location_type',value_name='distance')
```

```python
fasla_long=fasla.melt(id_vars='zip',var_name='location_type',value_name='distance')
```

```python
import seaborn as sns
```

```python
sns.barplot(x='zip',y='distance',hue='location_type',data=fasla_long)
```
