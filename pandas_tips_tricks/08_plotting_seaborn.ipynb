## Plotting-Seaborn

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Important things for plotting according to grammer of graphics
1. Data
2. Types of plot
3. y axis
4. x axis
5. Grouping(hue/color) based on categorical variables
6. Size of the plot
7. Color pallets of the plot
8. Style of the plot
9. How do you interpret


```python
# Data
df=sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# Scatter plot
# Figure size
plt.figure(figsize=(10, 5))
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
# Labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age and Fare')
#Save the plot
plt.savefig('./outputs/plots/scatter_plot.png',dpi=300)
plt.savefig('./outputs/plots/scatter_plot.svg',dpi=300)
```

```python
# Scatter plot
# Figure size
plt.figure(figsize=(10, 5))
# Change the scatter points color
palette={'male':'blue','female':'red'}
sns.scatterplot(x='age', y='fare', data=df, hue='sex', palette=palette)
```

```python
# Set the style
sns.set_theme(style='darkgrid',font_scale=1.5, font='Arial')


plt.figure(figsize=(10, 5))
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
# Labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age and Fare')
```

### Interpretation
1. We have age of the passengers at the x axis and the fare at the y axis.
2. The scatter plot shows the relationship between the age and fare of the Titanic passengers.
3. we futher grouped the passengers by their gendersuch as male(blue points) and the female(orange points).
4. Two male passengers paid the highest fare and they might be the first class passengers.
5. The oldest passenger of Titanic is also a male having 80 years of age.
