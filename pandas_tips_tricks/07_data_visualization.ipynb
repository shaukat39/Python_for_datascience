## Data Visualization


```python
# dataviz
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Import titanic data
df = sns.load_dataset('titanic')
```

```python
# To plot a scatter plot
df.plot(kind='scatter', x='age', y='fare', c='pclass', colormap='viridis', s=50, alpha=0.5)
```

```python
df.head()
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.ylabel('Fare')
plt.xlabel('Gender')
```

```python
# Bar plot using seaborn
sns.barplot(data=df, x='survived', y='fare')
```

```python
sns.barplot(data=df, x='sex', y='fare',hue='class')
```

```python
sns.barplot(data=df, x='sex', y='age',hue='class')
```

```python
sns.boxplot(data=df, x='sex', y='age',hue='class')
```

```python
sns.scatterplot(data=df, x='fare', y='age',hue='class')
```

```python
sns.scatterplot(data=df, x='fare', y='age',hue='sex')
```
