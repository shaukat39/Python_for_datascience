```python
df_01.head()
```

```python
# To remove the dollar sign from the Price column
df_01['Price']=df_01['Price'].apply(lambda x: str(x).replace('$',''))
```

```python
df_01.head(20)
```

```python
# To check a single column
df_01['Price']
```

```python
# To check the non zero values in the Price column
df_01[df_01['Price']!='0']
```

```python
# To conver the Price column to float
df_01['Price']=df_01['Price'].astype(float)
```

```python
# To check the missing values in the dataset
df_01.isnull().sum()
```

```python
# To convert KB to MB
df_01['Size']=df_01['Size'].apply(lambda x: str(x).replace('k','e-3'))
```

```python
df_01['Size']=df_01['Size'].apply(lambda x: str(x).replace('M',''))
```

```python
# To check the unique values in the Size column
df_01['Size'].unique()
```

```python
# To convert the string in size column to mean value
df_01['Size']=df_01['Size'].apply(lambda x: str(x).replace('Varies with device','NaN'))
```

```python
# To convert the Size column to float
df_01['Size']=df_01['Size'].astype(float)
```
