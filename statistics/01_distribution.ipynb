```python
import numpy as np
from scipy.stats import skew,kurtosis

np.random.seed(0)
data=np.random.normal(0,1,1000) # generate 1000 random numbers from a normal distribution with mean 0 and standard deviation 1
print("Skewness: ", skew(data))
print("Kurtosis: ", kurtosis(data))
```

```python
sns.histplot(data,kde=True)
```

```python
# how to test normal distribution of data

#H0: data is normally distributed
#H1: data is not normally distributed
#test statistic: Shapiro-Wilk test
#p-value: 0.0001
#Conclusion: reject H0, data is not normally distributed
#Test statistic: Shapiro-Wilk test
#p-value: 0.0001
#Conclusion: reject H0, data is not normally distributed
#Test statistic: Shapiro-Wilk test
#p-value: 0.0001
#Conclusion: reject H0, data is not normally distributed
```

```python
stats.shapiro(data)
print(f"Shapiro-Wilk test statistics: {stats.shapiro(data)[0]}")
print(f"Shapiro-Wilk test p-value: {stats.shapiro(data)[1]}")
```
