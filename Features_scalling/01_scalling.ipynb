### MIn-Max Scalling

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data={'values':[10,20,30,40,50]}
df=pd.DataFrame(data)
# Scale the data using MinMax scaler
scaler=MinMaxScaler()
df['Scaled_values']=scaler.fit_transform(df['values'].values.reshape(-1,1))
print(df)
```

2. Standard scaler or Z-score Normalization

```python
from sklearn.preprocessing import StandardScaler
data={'values':[10,20,30,40,50]}
df=pd.DataFrame(data)
# Scale the data using MinMax scaler
scaler=StandardScaler()
df['Scaled_values']=scaler.fit_transform(df['values'].values.reshape(-1,1))
print(df)
```

3. Robust scaler

```python
from sklearn.preprocessing import RobustScaler
data={'values':[10,20,30,40,50]}
df=pd.DataFrame(data)
# Scale the data using MinMax scaler
scaler=RobustScaler()
df['Scaled_values']=scaler.fit_transform(df['values'].values.reshape(-1,1))
print(df)
```

4. Logrithmic scalling

```python
import numpy as np
import pandas as pd
```

```python
data={'values':[10000,20000,30000,40000,50000]}
df=pd.DataFrame(data)

df['values_log']=np.log(df['values'])
df['values_log2']=np.log2(df['values'])
df['values_log10']=np.log10(df['values'])

df.head()
```

```python
import seaborn as sns
sns.boxplot(df,x='values')
```

```python
sns.boxplot(df,x='values_log')
```

```python
sns.boxplot(df,x='values_log2')
```

```python
sns.boxplot(df,x='values_log10')
```
