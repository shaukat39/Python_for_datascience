## Sunburst Plot to show heirarchy

```python
import plotly.express as px
import seaborn as sns

titanic=sns.load_dataset('titanic')

fig=px.sunburst(titanic,
                path=['sex','class','who','alive','alone'],
                values='survived',color='sex')
fig.show()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
import pandas as pd
import numpy as np
```

```python
# To make a sunburst plot on tips dataset
sunburst_plot = px.sunburst(
    tips,
    path=['total_bill', 'sex', 'smoker', 'day','time'],
    values='size',
    color='tip',
    title='Sunburst Plot of Total Bill',
    )
sunburst_plot.show()


```
